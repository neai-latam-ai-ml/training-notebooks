##Creando un modelo con parametros estaticos

```python
from torch import nn
```

```python
X = torch.tensor([
    [10.0],
    [38.0],
    [100.0],
    [150.0]
])

model = nn.Linear(1,1) # one input and one output
print(model)
print(model.bias)   #inicializado con numero random
print(model.weight) #inicializado con numero random
```

```python
y_pred = model(X)
print(y_pred)
```

Ajustar el **BIAS**
Cerca de los valores pero todavia falta el *weight*

```python
model.bias = nn.Parameter(
    torch.tensor([32.0])
)
y_pred = model(X)
print(y_pred)
```

AJUSTAR el **WEIGHT**
Llegamos a la misma prediccion pero sin hacer uso de la formula:

```python
model.weight = nn.Parameter(
    torch.tensor([[1.8]]) # Notice: X matrix has one row and one column
)
y_pred = model(X)
print(y_pred)
```
